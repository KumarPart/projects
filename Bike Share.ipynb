```python
# read in the csv file as a pandas dataframe
bikes = pd.read_csv('london_merged.csv')
```

```python
# specifying the col names as I want them
new_cols_dict = {
    'timestamp' : 'Time',
    'cnt' : 'Count',
    't1' : 'Temp_real_F',
    't2' : 'Temp_feels_like_F',
    'hum' : 'Humidity_percent',
    'wind_speed' : 'Wind_speed_mph',
    'weather_code' : 'Weather',
    'is_holiday' : 'is_holiday',
    'is_weekend' : 'is_weekend',
    'season' : 'Season'
}
```

```python
# renaming the cols to the specified names
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
#changing humidity values to percent
bikes.Humidity_percent = round(bikes.Humidity_percent/100, 2)
```

```python
# creating a season dictionary to change numbers to season name
season_dict ={
    '0.0' : 'Spring',
    '1.0' : 'Summer',
    '2.0' : 'Fall',
    '3.0' : 'Winter'
}
```

```python
# creating a weather dictionary to name weather instead of int
weather_dict = {
    '1.0' : 'Clear',
    '2.0' : 'Scattered Clouds',
    '3.0' : 'Broken Clouds',
    '4.0' : 'Cloudy',
    '7.0' : 'Rain',
    '10.0' : 'Rain with thunderstorm',
    '26.0' : 'Snowfall'
}
```

```python
# change the seasons col data type 
bikes.Season = bikes.Season.astype('str')

# mapping the values of int to str
bikes.Season = bikes.Season.map(season_dict)
```

```python
# change the weather col data type 
bikes.Weather = bikes.Weather.astype('str')

# mapping the values of int to str
bikes.Weather = bikes.Weather.map(weather_dict)
```

```python
# check dataframe
bikes.head()
```

```python
 # change from Cel to Far

bikes['Temp_real_F'] = bikes['Temp_real_F'].apply(lambda x: round(x * 1.8 + 32))
```

```python
bikes.head()
```

```python
# change from Cel to Far

bikes['Temp_feels_like_F'] = bikes['Temp_feels_like_F'].apply(lambda x: round(x * 1.8 + 32))
```

```python
bikes.head()
```

```python
# kph to mph 
bikes['Wind_speed_mph'] = bikes['Wind_speed_mph'].apply(lambda x: round(x *0.6214))
```
